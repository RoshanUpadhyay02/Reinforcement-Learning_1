```python
import numpy as np
import random

n_machines = 3
p = 10
coins = 20
r = ['7','B','BB','BBB','C','DD']

def puls(r):
  l1 = random.choice(r)
  l2 = random.choice(r)
  l3 = random.choice(r)
  if l1 == 'DD' and l2 == 'DD' and l3 == 'DD':
    reward = 60
  elif l1 == '7' and l2 == '7' and l3 == '7':
    reward = 50
  elif l1 == 'BBB' and l2 == 'BBB' and l3 == 'BBB':
    reward = 40
  elif l1 == 'BB' and l2 == 'BB' and l3 == 'BB':
    reward = 30
  elif l1 == 'B' and l2 == 'B' and l3 == 'B':
    reward = 20
  elif l1 == 'C' and l2 == 'C' and l3 == 'C':
    reward = 10
  else:
    reward = -10
  s = l1 + " " + l2 + " " + l3
  return s,reward


def pulls(k):
  if k == 1:
    r1 = []
    a1 = []
    for i in range(90):
      a,b = puls(r)
      r1.append(b)
      a1.append(a)
    m = random.choice([max(r1),min(r1)])
    print('Combination: ',a1[r1.index(m)])
    return m

  if k == 2:
    r2 = []
    a2 = []
    for i in range(95):
      a,b = puls(r)
      r2.append(b)
      a2.append(a)
    m = random.choice([max(r2),min(r2)])
    print('Combination: ',a2[r2.index(m)])
    return m

  if k == 3:
    r3 = []
    a3 = []
    for i in range(80):
      a,b = puls(r)
      r3.append(b)
      a3.append(a)
    m = random.choice([max(r3),min(r3)])
    print('Combination: ',a3[r3.index(m)])
    return m




def begin():
  p_r = 0
  print("Pick a Machine from: 1 , 2, 3")
  k = int(input('Which Machine you want to pick: '))
  p_r += pulls(k)

  return p_r

x = 0
z = 0
print("Welcome to the slot machine!")
while coins > 0 :
  print('Coins Left: ',coins)
  print()
  n = input('Do you want to spin? (y/n): ')
  if n == 'Y' or n == 'y':
    x = begin()
    print('Rewards earned in this round: ',x)
    z += x
    print("Total Rewards Obtined: ",z)
    print()
  elif n == "n" or n == 'N':
        break
  else:
      print("Invalid input. Please enter y or n.")
  coins -= 1
print()
print("Thanks for playing!\n\n\t||| Rewards at the end : {}. |||".format(z))
```
